```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mat = np.memmap("omp.out.dat", np.double, "r")
```

```python
mat = np.array(mat.reshape(10000 // 100, 1000))
```

```python
# Plotting each series
timesteps = np.arange(0, 10000, 100)  # Create an array of timesteps equal to the number of rows
for i in range(mat.shape[1] // 15):  # Loop over each column (data series)
    plt.plot(timesteps, mat[:, i], label=f'Series {i+1}')
plt.plot(timesteps, [104] * (10000 // 100), color="black")

# Adding title and labels
plt.title('Underlying Asset Price Over Time (OpenMP, 16 threads)')
plt.xlabel('Time (s)')
plt.ylabel('Underlying Price ($)')

# Show plot
plt.show()

plt.savefig("omp.png")
```
